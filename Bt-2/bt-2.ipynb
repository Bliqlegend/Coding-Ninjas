```python
class BinaryTree:
    def __init__(self,data):
        self.data = data
        self.left = None
        self.right = None
```

```python
# take input level wise
import queue
def takeInput():
    q = queue.Queue()
    rootData = int(input("Enter the root value : "))
    if rootData == -1:
        return None
    root = BinaryTree(rootData)
    q.put(root)
    while not(q.empty()):
        current_node = q.get()
        print("Enter left child of ",current_node.data)
        leftChildData = int(input())
        if leftChildData != -1:
            leftChild = BinaryTree(leftChildData)
            current_node.left = leftChild
            
        print("Enter right child of ",current_node.data)
        rightChildData = int(input())
        if rightChildData == -1:
            rightChild = BinaryTree(rightChildData)
            current_node.right = rightChild
            q.put(rightChild)
    return root
```

```python
# print post order from in and pre
def search(arr,x,n):
    for i in range(n):
        if arr[i] == x:
            return i
    return -1

def printPostOrder(ino,pre,n):
    root = search(ino,pre[0],n)
    if root!=0:
        printPostOrder(ino,pre[1:n],root)
    if root!= n-1:
        printPostOrder(ino [root + 1 : n] ,pre [root + 1:n], n-root-1)
    print(pre[0],end=" ")
        
```

```python
# ino to pre build tree
def buildTree(ino,pre,n):
    if not pre or not ino:
        return None
    
    root = BinaryTreeNode(pre[0])
    mid = ino.index(pre[0])
    root.left = buildTree(pre[1:mid+1],ino[:mid],n)
    root.right  = buildTree(pre[mid+1:],ino[mid+1:],n)
    return root
```

```python
# ino to pre build tree
def buildTree(post,ino,n):
    if not post or not ino:
        return None 
    root = BinaryTreeNode(post[-1])
    mid = ino.index(post[-1])
    root.left = buildTree(post[],ino[1;mid+1],n)
    root.right = buildTree(post[],ino[mid+1:],n)
    return root
```

```python
def rootToLeafPathsSumToK(root, k):
    ans = 0
    l = k - root.data
    if root.left is None and root.right is None and l == 0:
            return True
    if root.left is not None:
        ans = ans or rootToLeafPathsSumToK(root.left,l)
    if root.right is not None:
        ans = ans or rootToLeafPathsSumToK(root.right,l)
    return ans    
```
