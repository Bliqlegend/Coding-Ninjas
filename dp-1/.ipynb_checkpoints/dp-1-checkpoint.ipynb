```python
def fibb(n,dp):
    
    if n == 0 or n == 1:
        return n
    
    if dp[n-1] == -1:
        ans1 = fibb(n-1,dp)
        dp[n-1] = ans1
    else:
        ans1 = dp[n-1]
    
    if dp[n-2] == -1:
        ans2 = fibb(n-2,dp)
        dp[n-2] = ans2
    else:
        ans2 = dp[n-2]
    
    myAns = ans1 + ans2
    return myAns

def fibbI(n):
    dp = [0 for i in range(n+1)]
    dp[0] = 0
    dp[1] = 1
    i = 2
    while i<=n:
        dp[i] = dp[i-1] + dp[i-2]
        i+=1
    return dp[n]

n = int(input())
# dp = [-1 for i in range(n+1)]
ans = fibbI(n)
print(ans)
```

```python
from sys import setrecursionlimit
setrecursionlimit(10**6)
import sys, math

def minSquares(n,dp):
    if n==0:
        return 0
    
    root = int(math.sqrt(n))
    ans = sys.maxsize
    for i in range(1,root+1):
        if dp[n-(i**2)] == -1:
            currAns = 1 + minSquares(n-(i**2),dp)
            dp[n-(i**2)] = currAns
        else:
            currAns = dp[n-(i**2)]
        ans = min(ans,currAns)
    return ans

def minSquares(n):
    dp [-1 for i in range(n+1)]
    dp[0] = 0
    for i in range(1,n+1):
        root = int(math.sqrt(i))
        ans = sys.maxsize
        for j in range(1,root+1):
            currAns = 1 + dp[i-(j**2)]
            ans = min(currAns,ans)
        dp[i] = ans
    return ans
            
    
    
    
n = int(input())
dp = [-1 for i in range(n+1)]
ans = minSquares(n,dp)
print(ans)
```
